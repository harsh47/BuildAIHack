# Top Order Requests by ProductType

## Product Type Donations

```python
plt.figure(figsize=(8,6))
sns.countplot(x=orderItemsDonations.ProductType, palette='winter')
plt.xticks(rotation=90)
plt.show()
```

## Product Type Requests 

```python
plt.figure(figsize=(8,6))
sns.countplot(x=orderItemsRequests.ProductType, palette='winter')
plt.xticks(rotation=90)
plt.show()
```

## Product Type Donations

```python
plt.figure(figsize=(8,6))
sns.countplot(x=orderItemsDonations.ProductItem, palette='winter')
plt.xticks(rotation=90)
plt.show()
```

## Product Type Requests 

```python
plt.figure(figsize=(8,6))
sns.countplot(x=orderItemsRequests.ProductItem, palette='winter')
plt.xticks(rotation=90)
plt.show()
```
